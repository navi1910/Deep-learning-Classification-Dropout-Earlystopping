```python
# Imports 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('cancer_classification.csv')
```

```python
X = df.drop('benign_0__mal_1', axis = 1).values
y = df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , random_state = 101, test_size = 0.25)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))

model.add(Dense(15, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(x = X_train, y = y_train, epochs = 600,
          batch_size = 5, validation_data = (X_test, y_test), verbose = 0)
```

```python
history = pd.DataFrame(model.history.history)
history.plot()
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))

model.add(Dense(15, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
# Model with EarlyStopping
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', patience = 25, 
                          verbose = 1)
model.fit(x = X_train, y = y_train, epochs = 600,
          batch_size = 5, validation_data = (X_test, y_test), verbose = 1,
         callbacks = [early_stop])
```

```python
history = pd.DataFrame(model.history.history)
history.plot()
```

```python
# Model with Dropout Layers
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dropout(0.3))

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.3))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', patience = 25, 
                          verbose = 1)
model.fit(x = X_train, y = y_train, epochs = 600,
          batch_size = 5, validation_data = (X_test, y_test), verbose = 1,
         callbacks = [early_stop])
```

```python
history = pd.DataFrame(model.history.history)
history.plot()
```

```python
# Model with Dropout Layers
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dropout(0.3))

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.3))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', patience= 25,
                          verbose = 1)
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test, y_test),
          verbose = 1,callbacks = [early_stop])
```

```python
history = pd.DataFrame(model.history.history)
history.plot()
```

```python
y_pred = model.predict(X_test)
y_pred = np.round(y_pred).astype(int)
```

```python
print(confusion_matrix(y_pred, y_test))

print('\n')

print(classification_report(y_pred, y_test))
```
